```python
# Creates a list of ALL functions used in the proejct source and prints them
from os import listdir
from os.path import isfile, join
files = [f for f in listdir("../src/") if isfile(join("../src/", f))]
functions_to_test = []
for file in files:
    with open("../src/" + file, "r") as f:
        for line in f:
            if "def " in line:
                string = ""
                seen = 0
                found = False
                for char in line:
                    if char in ["d", "e", "f", " "] or found:
                        seen += 1
                    else:
                        seen = 0
                    if seen >= 4:
                        found = True
                        if char != "(":
                            string += char
                        else:
                            break
                if string[0] == "f":
                    string = string[1:]
                
                functions_to_test.append("src." + file[:-3] + "." + string.strip())
    functions_to_test.append("\n")

graphing_files = [f for f in listdir("../src/graphing") if isfile(join("../src/graphing", f))]
#graphing_functions_to_test = []
for file in graphing_files:
    with open("../src/graphing/" + file, "r") as f:
        for line in f:
            if "def " in line:
                string = ""
                seen = 0
                found = False
                for char in line:
                    if char in ["d", "e", "f", " "] or found:
                        seen += 1
                    else:
                        seen = 0
                    if seen >= 4:
                        found = True
                        if char != "(":
                            string += char
                        else:
                            break
                if string[0] == "f":
                    string = string[1:]
                
                functions_to_test.append("src.graphing." + file[:-3] + "."+ string.strip())
    functions_to_test.append("\n")

for line in functions_to_test:
    
    print(line)
```
