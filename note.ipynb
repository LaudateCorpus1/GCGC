```python
import numpy as np
random_numerators = np.random.rand(10)
random_denominators = np.random.rand(10)
print(random_numerators)
print(random_denominators)
random_numerators_sum = sum(random_numerators)
random_denominators_sum = sum(random_denominators)

print("Sums", random_numerators_sum / random_denominators_sum)

averages = []
for i in range(len(random_numerators)):
    val = random_numerators[i] / random_denominators[i]
    averages.append(val)

print("Avg", sum(averages) / len(averages))
```

```python
import numpy as np
import matplotlib.pyplot as plt

threshold = 0.1
data = np.random.random((10,10))
for row in range(10):
    for col in range(10):
        data[row][col] = round(data[row][col], 2)

fig, ax = plt.subplots()
im = ax.imshow(data, cmap=plt.cm.hot, interpolation='none', vmin=threshold)
cbar = fig.colorbar(im, extend='min')
cbar.cmap.set_under('green')

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(10):
    for j in range(10):
        text = ax.text(j, i, data[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
